**Criando colunas de de valores mensais e anuais**

#colunas numericas (Valores gastos por mês)

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
```

```python
dados = pd.read_csv(url, sep=';')
dados.head()
```

```python
dados['valor_por_mês'] = dados['Valor'] + dados['Condominio']
dados.head(10)
```

# Tratando as linhas com valores Nan

```python
dados.fillna(0, inplace=True)
dados['valor_por_mês'] = dados['Valor'] + dados['Condominio']
dados.head()
```

#Criando a coluna de valores anuais

```python
dados['Valor_por_ano'] = dados['valor_por_mês'] * 12 + dados['IPTU']
dados.head(20)
```

#** Criando colunas categoricas **

```python
dados['Descrição'] = dados['Tipo'] + dados['Bairro']
dados.head()
```

```python
dados['Descrição'] = dados['Tipo'] + ' em ' + dados['Bairro']
dados.head()
```

```python
dados['Descrição'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                                        dados['Quartos'].astype(str) + ' Quartos(s) ' + \
                                        ' e ' + dados['Valor'].astype(str)  + ' vagas de garagem. '
dados.head()
```

**Coluna binaria**

# Mostrando quais imoveis possuem suites

```python
dados['Possui_suite'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else "Não")
dados.head()
```

```python
dados.to_csv('dados_completos_dev.csv', index=False, sep=';')
```
